```python
# Run it.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from numpy import random;from random import randint
N =randint(10, 2150)
url='https://raw.githubusercontent.com/edelmanarik/DataVisualFri/main/TestBase_02032022.csv'
DataTest = pd.read_csv(url)
NameToChang=DataTest['OrderID'][N]
DataTest.loc[DataTest['OrderID']==NameToChang,'OrderID']=IdNumber
DataTest.loc[DataTest['OrderID']==IdNumber,'Cutomer Name']=MyName
DataTest.loc[DataTest['OrderID']==IdNumber,'Customer Country']='USA'
DataTest.sample(4)
```

### Question 2: Create columns 'Cost', 'New_Fee', 'All_Costs'

```python
DataTest['Sales Cost']=DataTest['Sales Cost'].astype(float)
```

```python
DataTest['Cost'] = DataTest['Sales Cost'] * DataTest['Sales Quantity']

DataTest.head()
```

```python
DataTest['New_Fee'] = DataTest['Fee'] / 100

DataTest.head()
```

```python
DataTest['Discount_%'] = DataTest['Discount'] / 100

DataTest.head()
```

```python
DataTest['All_Costs'] = (DataTest['Cost'] + DataTest['New_Fee']) * (1 - DataTest['Discount_%'])

DataTest.head()
```

### Question 3: Create columns 'Day_Name', 'Month_Name', 'Year'

```python
import datetime as dt
```

```python
DataTest['OrderDate'] = pd.to_datetime(DataTest['OrderDate'])

DataTest['Day_Name'] = DataTest['OrderDate'].dt.day_name()

DataTest['Month_Name'] = DataTest['OrderDate'].dt.month_name()

DataTest['Year'] = DataTest['OrderDate'].dt.year # attention without()

DataTest.head() 
```

### Question 12: A Heat Map must be built, showing the number of items sold, by department and day of the week, only for sales whose commissions are 800 or more.

```python
#df_for_heatmap_1 = DataTest[DataTest['Fee'] >= 800]

#df_for_heatmap_1.groupby(['Department', 'Day_Name'])

group = DataTest[DataTest['Fee'] >= 800].groupby(['Department', 'Day_Name']).size().unstack()

group
```

```python
sns.heatmap(group, annot=True, fmt = 'd')

# "annot=True" is showing the details (numbers) 

plt.title('Sold Items by Department, Day, with Fee above 800')

plt.figure(figsize = (16,12))

plt.show()
```
